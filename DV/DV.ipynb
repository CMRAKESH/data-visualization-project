```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")  # optional for nicer plots

```

```python
df = sns.load_dataset('iris')  # built-in dataset
df.head()  # preview first 5 rows
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.hist(df['sepal_length'], bins=10, color='skyblue', edgecolor='black')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df, hue='species', palette='Set2', dodge=False)
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df, s=100)
plt.show()
```

```python
sns.pairplot(df, hue='species', palette='bright')
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')  # selects only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df_encoded = df.copy()
df_encoded['species'] = df_encoded['species'].map({'setosa':0, 'versicolor':1, 'virginica':2})
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df, s=100)
plt.title('Sepal Length vs Petal Length')
plt.savefig('scatter_plot.png')  # saves to the current folder
plt.show()
```
